```python
import pandas as pd
from functools import reduce

# List of file paths for your CSV files
file_paths = ['2.csv', '4.csv', '5.csv','Weather.csv']  # Add your file paths here

# Read each CSV file into a DataFrame and store them in a list
dfs = [pd.read_csv(file) for file in file_paths]

# Convert the "Time" column to datetime format for each DataFrame
for df in dfs:
    df['Time'] = pd.to_datetime(df['Time'])
    df['Time'] = df['Time'].dt.strftime('%m/%d/%Y %H:%M')

# Print column names of each DataFrame
#for df in dfs:
#    print(df.head)
```

```python
# Merge DataFrames based on the "Time" column using reduce and a lambda function
#dfs.append(dfw);
merged_df = reduce(lambda left, right: pd.merge(left, right, on='Time', how='outer'), dfs)

```

```python
# Sort the DataFrame based on the "Time" column
merged_df.sort_values(by='Time', inplace=True)

```

```python
# To fill missing data with a specific value, we use the fillna method
# We fill missing values with 0
merged_df.fillna(0, inplace=True)

# To forward fill missing values, we use the ffill method
# merged_df.ffill(inplace=True)

# To interpolate missing values, we use the interpolate method
# merged_df.interpolate(inplace=True)

# Remove rows with missing values
# merged_df.dropna(inplace=True)
```

```python
# Save the merged DataFrame to a new CSV file
merged_df.to_csv('merged_data_final.csv', index=False)
```
